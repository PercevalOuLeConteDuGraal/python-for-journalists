# Ein erster Blick auf den Datensatz

Wenn wir einen neuen Datensatz bekommen, müssen wir uns zuerst darin zurecht finden, in diesem Notebook sehen wir, worauf wir bei der ersten Orientierung im Datensatz achten.

Zuerst schauen wir uns die grundlegende Struktur des Datensatzes an – was ist alles enthalten, wie heißen die Spalten wie groß ist der Datensatz?

## Datensatz reduzieren

Bei so großen Datensätzen ist es eine gute Idee, ihn auf die Spalten zu reduzieren, die für uns relevant sind. Was relevant ist, hängt von den Recherchefragen ab die wir haben. 

Bei diesem Datensatz können wir uns zum Beispiel dafür interessieren wie alt Hunde und Halter\*innen sind, welche Rasse und Farbe die Hunde haben und in welchen Vierteln sie wohnen.

Um den Datensatz auf die für uns wichtigen Spalten zu begrenzen ist es eine gute Idee, sich zuerst alle Spaltennamen ausgeben zu lassen

```python
dogs.columns
```

Die Spaltennamen können wir dann kopieren und alle Spalten entfernen, die uns nicht interessieren:

```python
dogs = dogs[['StichtagDatJahr', 'HalterId', 'AlterV10Cd', 'SexLang', 'KreisCd', 'QuarCd', 'QuarLang', 'Rasse1Text',
       'GebDatHundJahr', 'AlterVHundCd', 'SexHundLang', 'HundefarbeText']]

dogs
```

Damit ist der Datensatz schon ein bisschen übersichtlicher. Als nächstes schauen wir uns die einzelnen Spalten genauer an, ob die Inhalte in den Spalten realistisch klingen oder ob bei der Datenerhebung Fehler unterlaufen sind.

**Dieser Schritt ist einer der wichtigsten in jeder Datenrecherche: Wenn der analysierte Datensatz Fehler enthält, dann werden auch die Ergebnisse falsch sein.** Falsche Ergebnisse in der Datenanalyse sind genauso schlimm wie falsche O-Töne oder verdrehte Fakten in der Berichterstattung.

Zuerst überprüfen wir mit der `isna()`-Methode ob Felder in unserem Datensatz keinen Eintrag haben. Pandas hat dafür einen eigenen Datentyp: `pd.NA`. Dieser wird beim Import automatisch vergeben, wenn Felder in der Tabelle leer sind.

Damit die Tabelle übersichtlicher wird, summieren wir die Spalten auf.

```python
dogs.isna().sum()
```

Wir sehen, dass in unserem Datensatz keine Felder fehlen.

Für die weitere Analyse müssen wir in zwei Arten von Daten unterscheiden: Kategorielle und Numerische Daten.

## Kategorielle Daten

Kategorielle Daten sind Daten, die eine Zeile einer gewissen Kategorie zuordnen. Ein Beispiel in unserem Datensatz hier ist etwa das Geschlecht. Die Spalte `SexHundLang` drückt aus, welches Geschlecht der Hund in der entsprechenden Zeile hat. Meine Erwartung an den Datensatz ist, dass die Hunde nur in männlich und weiblich kategorisiert werden.

Mit der Methode `value_counts()` können wir uns anzeigen lassen, welche Kategorien wie häufig in dem Datensatz vorkommen:

```python
dogs["SexHundLang"].value_counts()
```

Wir sehen, dass tatsächlich nur die beiden Geschlechter vorkommen. Und auch die Verteilung entspricht meiner Erwartung: Männliche und weibliche Hunde sind etwa 50/50 verteilt.

Bei den Halter\*innen gehe ich davon aus, dass die Verteilung ähnlich ist:

```python
dogs[["SexLang"]].value_counts()
```

Hier sehen wir, dass es einen Überhang hin zu Frauen gibt. Ich habe nicht erwartet, dass zwei Drittel der Halter\*innen Frauen sind. Um sicher zu gehen, schauen wir uns deshalb an, wie viele Halter\*innen geführt sind, und ob die Zahl der Gesamtzahl entspricht:

```python
dogs["HalterId"].size
```

```python
dogs[["SexLang"]].value_counts().sum()
```

So sehen wir, dass alle Halter\*innen erfasst sind und anscheinend wirklich viel mehr Frauen als Männer Hunde halten.

Jetzt werfen wir einen Blick auf die Rasse des Hundes:

```python
dogs["Rasse1Text"].value_counts()
```

Wir sehen, dass 384 verschiedene Rassen aufgeführt sind. Eigentlich sind es nur 383, weil die häufigste Rasse "Unbekannt" ist. Das merken wir uns für später, wenn wir den Datensatz bereinigen.

Allgemein ist diese Ansicht etwas unübersichtlich. Deshalb lassen wir sie uns schnell als Graph ausgeben.

Dafür wandeln wir die Ausgabe von `value_counts()` wieder in einen `DataFrame` um und rufen die `.plot()` Methode auf. Diese Methode gibt uns dann ein einfaches Diagramm zurück, aus dem wir die Verteilung der Daten sehen können.

Mit dem Parameter `kind="bar"` in der `plot()`-Methode geben wir an, dass wir einen Barchart, also ein Balkendiagramm haben wollen. Normalerweise würde ein Liniendiagramm zurückgegeben – Liniendiagramme können wir aber nur für numerische Daten verwenden. Mehr dazu im Video zu Visualierung.

```python
pd.DataFrame(dogs["Rasse1Text"].value_counts()).plot(kind="bar")
```

Auf diesem Plot erkennt man nicht viel, aber wir sehen die grundlegende Verteilung: Es gibt einige Hunderassen die sehr häufig vorkommen und viele die eher selten sind. Das deckt sich mit meiner Erwartungshaltung.

## Numerische Daten

Numerische Daten sind einfach gesagt Zahlen. Dabei können wir zwischen steten und unsteten Verteilungen unterscheiden. 

Stetige Verteilungen sind zum Beispiel Zeitdaten. Wenn wir jede Minute am Tag die Temperatur vor unserer Tür messen, könenn wir eine Linie über den Tag zeichnen, die anzeigt wie sich die Temperatur entwickelt hat. 

Eine Unstete Verteilung ist etwa das Alter der Hunde in unserem Datensatz. Hier sind nur ganze Jahre angegeben, wir könenn also keinen Verlauf zeigen sondern, wie bei den kategoriellen Daten, nur angeben wie viele Hunde ein gewisses Alter haben.

Sehen wir uns die Verteilung des Alters der Hunde genauer an, wieder mit der `value_counts()` Methode:

```python
dogs["AlterVHundCd"].value_counts()
```

Hier fällt sofort auf, dass 8 Hunde angeblich 999 Jahre alt sind. Das ist natürlich Quatsch und heißt vermutlich, dass für das Alter des Hundes nichts eingegeben wurde, oder dass das Alter des Hundes nicht bekannt ist.

Auch das merken wir uns für die Datenbereinigung später, damit wir diese fehlerhaften Spalten mit `pd.NA` ersetzen können.

Sonst zeigt sich ein ähnliches Bild: Viele Hunde sind sehr jung und mit fortschreitendem Alter werden es weniger Hunde. 23 ist zwar sehr alt für einen Hund, aber er ist damit immer noch jünger als [der älteste Hund der Welt](https://www.nordbayern.de/panorama/bobi-ist-mit-30-jahren-der-alteste-hund-der-welt-sein-besitzer-verrat-sein-geheimnis-1.12992998).

Wenn wir das Geburtsjahr der Hunde überprüfen ergibt sich für uns das gleiche Bild:

```python
dogs["GebDatHundJahr"].value_counts()
```

8 Hund sind im Jahr 9999 geboren, was wir später ebenfalls durch den nicht vorhanenden Wert `pd.NA` ersetzen.

Hier fällt uns noch etwas auf: Der Datensatz ist von 2022 und der älteste Hund ist im Jahr 1995 geboren, er müsste also 28 Jahre alt sein. Aber laut der Spalte Alter ist der älteste Hund nur 23. 

Deshalb können wir vermuten, dass es in Zürich gar keine 60.000 Hunde gibt, sondern jeder Hund in jedem Jahr aufgeführt wird.

Sehen wir uns deshalb genauer an, wie oft der 23 jährige Hund in unserem Datensatz auftaucht. Leider haben die Hunde keine ID bekommen, sondern nur ihre Herrchen und Frauchen.

Also finden wir zuerst die Halter-ID des ältesten Hundes:

```python
dogs[dogs["AlterVHundCd"] == 23]
```

Die HalterId ist also `84494`. Mit einer ähnlichen Abfrage können wir nun alle Jahre abfragen in dem der Hundehalter in der Statistik auftaucht:

```python
dogs[dogs["HalterId"] == 84494]
```

Mit dieser Abfrage können wir sicher sein, dass der gleiche Hund mehrfach im Datensatz auftaucht und wir nicht einfach die Zeilen zusammenzählen können.

Diese Information stand nicht in der Beschreibung des Datensatzes - wir sehen also wie wichtig es ist, sich ein genaues Bild vom Datensatz zu machen, um nicht aus Versehen falsche Zahlen zu veröffentlichen.
