```python
import pandas as pd

filename = "/content/BEV326T3260_Bevoelkerung_nach-Geschlecht-Stadtkreis-Stadtquartier.xlsx"
population = pd.ExcelFile(filename)
```

```python
population.sheet_names
```

```python
pop21 = pd.read_excel(filename, sheet_name="2021")
pop21
```

```python
pop21 = pop21.drop(range(0,9)).drop(["Unnamed: 2", "Unnamed: 3"], axis=1)
pop21
```

```python
pop21.columns = ["Gebiet", "Total"]
pop21
```

```python
pop21["Jahr"] = 2021
pop21
```

```python
dfs = []

for year in population.sheet_names[1:]:
  dfi = pd.read_excel(filename, sheet_name=year)
  dfi = dfi.drop(range(0,9)).drop(["Unnamed: 2", "Unnamed: 3"], axis=1)
  dfi.columns = ["Gebiet", "Total"]
  dfi["Jahr"] = int(year)

  dfs.append(dfi)
```

```python
population = pd.concat(dfs)
```

```python
population = population.reset_index(drop=True)
population
```

```python
population.to_csv("population_zurich.csv")
```
